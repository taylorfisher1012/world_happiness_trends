Change it to a useable form (dataframe)

```python
import pandas as pd
```

```python
days_off = pd.read_html(str(soup.find('table', attrs={'class' : 'wikitable'})))[0]
days_off
```

```python
days_off = days_off.rename(columns = {'Country':'Country', 'Minimum annual leave':'Minimum annual leave', 'Paid vacation days by year (five-day-workweek)[1][2]':'Paid vacation days by year', 'Paid public holidays[3][4]':'Paid public holidays','Total paid leave (five day workweek)':'Total paid leave'})
```

```python
days_off
```

```python
days_off.dtypes
```

```python
pd.set_option('display.max_rows', None)
```

```python
days_off
```

```python
days_off['Total paid leave (five-day workweek)'] = days_off['Total paid leave (five-day workweek)'].str[:2]
days_off
```

```python
days_off['Total paid leave (five-day workweek)'] = days_off['Total paid leave (five-day workweek)'].dropna().astype(str).astype(int)
```
